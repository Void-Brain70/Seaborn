Distplots in seaborn

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p = sns.load_dataset("penguins")
p.head()
```

```python
#drawing a simple distplot
sns.distplot(p.flipper_length_mm)
```

```python
sns.distplot(p.flipper_length_mm,rug=True)
```

```python
#Removing the KDE plot
sns.distplot(p.flipper_length_mm,kde=False) #hist=Flase
```

```python
# Setting the number of bins
sns.distplot(p.flipper_length_mm,bins=20)
```

```python
# Setting the bin values as array
sns.distplot(p.flipper_length_mm,bins=[160,170,180,190,200,240])
```

```python
# Fitting the data to a particular distribution using norm_hist
sns.distplot(p.flipper_length_mm,norm_hist=True)
```

```python
sns.distplot(p.flipper_length_mm,norm_hist=True,kde=False)
```

```python
# Styling Rugplots
sns.distplot(p.flipper_length_mm, rug_kws={'color':'red','height':0.30},rug=True)
```

```python
# Styling Histograms
sns.distplot(p.flipper_length_mm, hist_kws={'color':'blue','alpha':0.3,'histtype':'step'},kde=False)
```

```python
sns.distplot(p.flipper_length_mm, hist_kws={'color':'blue','alpha':0.3,'histtype':'stepfilled'},kde=False)
```

```python
# Styling KDE plots
sns.distplot(p.flipper_length_mm,kde=True,hist=False,kde_kws={'color':'red','lw':2.5,'alpha':0.4})
```
